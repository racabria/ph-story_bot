```python
import logging
import sys
import spacy
from telegram.ext import Updater, CommandHandler, MessageHandler, Filters, ConversationHandler
import wikipedia
import pandas as pd
import datetime
import re
import ast

# LOAD PHILIPPINE HISTORY TIMELINE
# Source: https://en.wikipedia.org/wiki/Timeline_of_Philippine_history
file = open("ph_history_timeline.txt", "r")
contents = file.read()
timeline_dict = ast.literal_eval(contents)
file.close()

# LOAD TODAY IN HISTORY DATA
# Source: http://philippines-islands-lemuria.blogspot.com/2011/12/1-december.html
file = open("today_in_history.txt", "r")
contents = file.read()
tih_dict = ast.literal_eval(contents)
file.close()

logger = logging.getLogger(__name__)
logging.basicConfig(stream = sys.stdout, level=logging.DEBUG)

TELEGRAM_TOKEN = "" # YOUR TELEGRAM TOKEN HERE

def start(update,context):
    update.message.reply_text('Hi! Welcome to PH-story! Philippine History at your fingertips!')
    today = update.message.date.strftime('%m-%-d')
    if tih_dict[today]:
        update.message.reply_text('Today in Philippine History! ' + today)
        update.message.reply_text(tih_dict[today])
    return 'SEARCHING'

def wiki(update,context):
    msg = update.message.text

    try:
        year = re.search(r"(\d{4})", msg).group(1)
        if year in timeline_dict.keys():
            update.message.reply_text(year + ": " + timeline_dict[year])
    except:
        result = ''

    search = contextualize(msg)
    try:
        result = wikipedia.summary(search, sentences=2)
        update.message.reply_text(result)
    except wikipedia.exceptions.DisambiguationError as e:
        result = "Do you mean any of these? " + str(e.options[0:3])
        update.message.reply_text(result)
    except wikipedia.exceptions.PageError:
        result = search + " does not match any pages. Try another query!"
        update.message.reply_text(result)
    except wikipedia.exceptions.WikipediaException:
        result = "Wikipedia Unknown error."
        update.message.reply_text(result)

    result = wikipedia.search(search, results=3)
    if result:
        update.message.reply_text("Other Results: " + str(result))

    return 'SEARCHING'


def contextualize(search):
    return search + " History Philippines" # Philippine History

def add_info(update,context):
    return update.message.text

def today_in_history(date):
    return tih_dict[date]
    
def cancel(update,context):
    update.message.reply_text('Have a nice day!')
    return ConversationHandler.END

def main():
    updater = Updater(TELEGRAM_TOKEN, use_context = True)
    disp = updater.dispatcher
    conv_handler = ConversationHandler(
        entry_points = [ CommandHandler('start',start) ],
        states = {
            'SEARCHING': [ MessageHandler(Filters.text, wiki) ],
            'ADD_INFO': [ MessageHandler(Filters.text, add_info) ],
        },
        fallbacks = [ CommandHandler('cancel',cancel) ]
    )
    disp.add_handler(conv_handler)
    updater.start_polling()
    updater.idle()
    
if __name__ == '__main__':
    main()
```
